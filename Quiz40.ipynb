```python
# Quiz 40
def matches(l, s):
  time=l*100/s
  number_of_matches=time/5
  return (number_of_matches)

#tests
print(int(matches(100, 100)))
print(int(matches(250, 110)))
print(int(matches(500, 150)))
print(int(matches(12345, 123)))
```
