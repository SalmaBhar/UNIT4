```python
#Quiz 38
def givesize(data :list):
  maximum=max(data)
  sum=0
  spaces=0
  for n in range(0, len(data)):
    while sum<=maximum:
      sum+=data[n]
    spaces+=1
  return (spaces*sum)

#tests
print(givesize([1,2,4,10,2,4,2,16,2,1,1]))
print(givesize([64,1]))
print(givesize([4,3,3,3,3,3]))
```
