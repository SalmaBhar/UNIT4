# **Quiz 39**

```python
#Quiz 39
def narcissistic(m, n):
  nar_list=[]
  digits=[]
  sum=0
  for num in range(m+1, n):
    digits=str(num)
    for d in range(0, len(digits)):
      sum+=digits[d]*len(digits)
    if sum==num:
      nar_list.append(num)
  return (len(nar_list))

#tests
print(narcissistic(0, 100))
print(narcissistic(4, 59))
print(narcissistic(3, 6))
```
