# **Quiz 43 Code**

```python
# program
def rectangle(h, w, c, cv, ch):
  first_last= f"{c}{ch*(w-2)}{c}"
  print(first_last)
  for i in range(h-2):
    middle= f"{cv}{' ' *(w-2)}{cv}"
    print(middle)
  print(first_last)

#testing
rectangle(5, 5, "*", "|", "-")
rectangle(5, 6, "0", "i", "_")
```
